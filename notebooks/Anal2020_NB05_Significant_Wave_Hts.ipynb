```python
plot_path = '../../../miscellaneous/Webinar Nov 2020/'
```

```python
# THE DATA (NOTE THAT WAVE HEIGHTS ARE APPROXIMATE)

months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
wave_heights = [3.6,3.8,3.8,3.45,3.05,2.45,2.1,2.7,3.6,3.9,3.55,3.7]         # these values approximated from chart...
```
